# 画像の読み込みと表示

```
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

print(cv2.__version__) 
```

## ファイルアップロードと描画

cameraman.jpgというファイルをgoogle colabにアップロードします。

アップロードしたら、読み込んでpyplot用にRGB変換してから表示します。

```
from google.colab import files
f = files.upload()
```

```
img=cv2.imread('./cameraman.jpg')

show_img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(show_img)
```

## フィルタ処理

```
gb_img=cv2.GaussianBlur(img,(5,5),0)
show_img=cv2.cvtColor(gb_img, cv2.COLOR_BGR2RGB)
plt.imshow(show_img)
```

## ダウンサンプリング

```
d1_img=cv2.pyrDown(img)
d2_img=cv2.pyrDown(d1_img)
plt.subplot(131)
show_img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(show_img)
plt.subplot(132)
show_img=cv2.cvtColor(d1_img, cv2.COLOR_BGR2RGB)
plt.imshow(show_img)
plt.subplot(133)
show_img=cv2.cvtColor(d2_img, cv2.COLOR_BGR2RGB)
plt.imshow(show_img)
```

## Canny

```
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_cny=cv2.Canny(gray,100,200)
plt.subplot(121)
plt.imshow(gray)
plt.subplot(122)
plt.imshow(img_cny)
```

```
# ダウンサンプリングするとノイズを除去できる
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray=cv2.pyrDown(gray)
img_cny=cv2.Canny(gray,100,200)
plt.subplot(121)
plt.imshow(gray)
plt.subplot(122)
plt.imshow(img_cny)
```

```
# フィルタ処理でもノイズを除去できる
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray=cv2.GaussianBlur(gray,(5,5),0)
img_cny=cv2.Canny(gray,100,200)
plt.subplot(121)
plt.imshow(gray)
plt.subplot(122)
plt.imshow(img_cny)
```

```
# フィルタ処理でもノイズを除去できる
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray=cv2.GaussianBlur(gray,(5,5),0)
gray=cv2.pyrDown(gray)
img_cny=cv2.Canny(gray,100,200)
plt.subplot(121)
plt.imshow(gray)
plt.subplot(122)
plt.imshow(img_cny)
```
